```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
out_0 = pd.read_csv('out_0.csv', header=None)
out_0.columns = ['time', 'rank']
```

```python
plt.figure(figsize=(12,8))
for rank, group in out_0.groupby('rank'):
    plt.plot(group.time, '.', label=rank)
plt.legend(fontsize=14)
plt.xlabel('mpi-recieve', fontsize=14)
plt.ylabel('time in seconds', fontsize=14)
plt.title('Times of message-recieves by process 0 from other processes', fontsize=16)
plt.savefig('Scheduler_times.pdf', format='pdf')
plt.show()
```

```python
ranks = np.unique(out_0['rank'])
plt.figure(figsize=(12,8))
for rank in ranks:
    data = pd.read_csv(f'out_{rank}.csv', header=None)
    plt.plot(data[0], '.--', label=rank)
plt.legend(fontsize=14)
plt.xlabel('mpi-recieves', fontsize=14)
plt.ylabel('time in seconds', fontsize=14)
plt.title('Times of message-recieves by processes other than 0', fontsize=16)
plt.savefig('Workers_times.pdf', format='pdf')
plt.show()
```

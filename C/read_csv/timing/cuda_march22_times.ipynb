```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
add_vectors = {
    "size": 12 + np.arange(12, 18),
    "cpu <<<2,  64>>>": [0.535, 0.837, 1.474, 2.614, 5.368, 10.592],
    "cpu <<<1, 128>>>": [0.526, 0.846, 1.488, 2.711, 4.919, 10.503],
    "cpu <<<2, 128>>>": [0.482, 0.784, 1.340, 2.329, 4.871,  9.515],
    "gpu <<<2,  64>>>": [0.385, 0.555, 0.879, 1.438, 2.535,  4.751],
    "gpu <<<1, 128>>>": [0.353, 0.561, 0.883, 1.433, 2.545,  4.772],
    "gpu <<<2, 128>>>": [0.322, 0.455, 0.634, 1.127, 1.925,  3.465]
}

add_vectors_df = pd.DataFrame(add_vectors)
```

```python
ax_0 = add_vectors_df.plot(x="size", y=add_vectors_df.columns.drop("size"), figsize=(12,8))
ax_0.set_title('Vector plus vector', fontsize=20)
ax_0.set_xlabel('$\log_2($vector size$)$', fontsize=16)
ax_0.set_ylabel(f'Time in seconds', fontsize=16)
ax_0.legend(fontsize=16)
plt.show()
```

```python
mult_matrix = {
    "size": np.arange(12, 18),
    "cpu <<<2,  64>>>": [2.746, 5.364, 10.520, 20.917, 41.615, 83.021],
    "cpu <<<1, 128>>>": [2.743, 5.346, 10.517, 20.911, 41.616, 83.192],
    "cpu <<<2, 128>>>": [2.726, 5.314, 10.462, 20.761, 41.424, 86.612],
    "gpu <<<2,  64>>>": [0.253, 0.281, 0.390, 0.454, 0.877, 1.432],
    "gpu <<<1, 128>>>": [0.234, 0.298, 0.385, 0.463, 0.884, 1.432],
    "gpu <<<2, 128>>>": [0.230, 0.259, 0.321, 0.432, 0.627, 1.001]
}

mult_matrix_df = pd.DataFrame(mult_matrix)
```

```python
ax_1 = mult_matrix_df.plot(x="size", y=mult_matrix_df.columns.drop("size"), figsize=(12,8))
ax_1.set_title('Matrix ($2^{dim_0} \otimes 2^{12}$) times vector', fontsize=20)
ax_1.set_xlabel('$\log_2(dim_0)$', fontsize=16)
ax_1.set_ylabel(f'Time in seconds', fontsize=16)
ax_1.legend(fontsize=16)
plt.show()
```
